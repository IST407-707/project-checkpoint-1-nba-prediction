```python
ad_stats_14 = pd.read_csv("/workspaces/project-checkpoint-1-nba-prediction/dev/advanced_stats_14_15.csv")
ad_stats_15 = pd.read_csv("/workspaces/project-checkpoint-1-nba-prediction/dev/advanced_stats_15_16.csv")
ad_stats_14 = ad_stats_14.rename(columns= {"BPM▼" : "BPM"})
ad_stats_15 = ad_stats_15.rename(columns= {"BPM▼" : "BPM"})
ad_stats_14["season"] = "2014-2015"
ad_stats_15["season"] = "2015-2016"
```

```python
#create a dictionary with season team and per list 
#seasons_bpm = pd.read_csv('seasons_per_s.csv')
teams = test["Tm"].unique()
#season,team_id,player_dir,per

def get_top_n_pers(seasons_pers, n): # can adjust to get pers of past seasonm on the new seasons roster
    roster_pers = {}
    for season in seasons_pers['season'].unique():
        subset = seasons_pers[seasons_pers['season'] == season]
        roster_pers[season] = {}
        for team in teams:
            team_subset = subset[subset['Tm'] ==  team]
            team_subset = team_subset.sort_values(by='BPM', ascending=False)
            roster_pers[season][team] = np.float16(team_subset['BPM'].iloc[:n]).flatten()
    return roster_pers

rosters = get_top_n_pers(test, 8)  # rosters[season][team]
```
